```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
region_Q_single=np.load('0.15_alpha_1.5_wta_1.0_wtp_1.0_Region_Q_single.npy')
region_Q_expert=np.load('0.15_Region_Q_ic.npy')
region_Q=np.load('0.15_alpha_2.0_wta_1.0_wtp_1.0_Region_Q.npy')
num=np.load('num_region.npy')
expert=np.mean(region_Q_expert,axis=0)
single=np.mean(region_Q_single,axis=0)
hi_rice=np.mean(region_Q,axis=0)

indices = np.arange(656)
bar_width = 1
palette = ['#D67C7C', '#7C9ED6', '#89AB89']  # Light red, light blue, brighter light green
darker_palette = ['#A55252', '#5273A5', '#657D65'] 

# Create figure and axes
fig, ax = plt.subplots(figsize=(25, 8), dpi=300)

# Plot bars for each group
ax.bar(indices, expert, width=bar_width, label='Expert', color=palette[0])
ax.bar(indices + 670, single, width=bar_width, label='Single-Agent', color=palette[1])
ax.bar(indices + 1340, hi_rice, width=bar_width, label='Hi-Rice', color=palette[2])
ax.set_xlim(0, 656 + 1340)

# Add dashed lines to separate the groups
plt.axvline(x=663, color='black', linestyle='--')
plt.axvline(x=1333, color='black', linestyle='--')

# Customize y-axis
plt.ylabel('The Intervention Cost \n of Each Region ($Q_i$)', size=22)
plt.yticks(size=20)

# Customize x-axis
plt.xlabel('Region Index', size=22)
plt.xticks(
    np.concatenate([
        np.arange(0, 656, 100), 
        np.arange(670, 670 + 656, 100), 
        np.arange(1340, 1340 + 656, 100)
    ]), 
    labels=np.concatenate([
        np.arange(0, 656, 100), 
        np.arange(0, 656, 100), 
        np.arange(0, 656, 100)
    ]), 
    fontsize=20
)

# Adding group labels with darker colors
plt.text(656/2, max(np.concatenate([expert, single, hi_rice])) * 1.1, 'Individual-Control', horizontalalignment='center', size=24, color='black')
plt.text(670 + 656/2, max(np.concatenate([expert, single, hi_rice])) * 1.1, 'Single-Agent RL', horizontalalignment='center', size=24, color='black')
plt.text(1340 + 656/2, max(np.concatenate([expert, single, hi_rice])) * 1.1, 'Hi-RICE', horizontalalignment='center', size=24, color='black')

# Remove legend
plt.legend().set_visible(False)
plt.savefig("ab_region_q1.png")
plt.show()

# 示例数据
indices = np.arange(656)
bar_width = 1
palette = ['#D67C7C', '#7C9ED6', '#89AB89']  # Light red, light blue, brighter light green
darker_palette = ['#A55252', '#5273A5', '#657D65'] 

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(25, 8), dpi=300)

# 示例数据替换为实际数据计算


expert = np.mean(np.exp(region_Q_expert / num), axis=0)
single = np.mean(np.exp(region_Q_single / num), axis=0)
hi_rice = np.mean(np.exp(region_Q / num), axis=0)

# 限制最大值为 10000
expert[expert > 10000] = 10000
single[single > 10000] = 10000
hi_rice[hi_rice > 10000] = 10000

# 绘制每组的柱状图
ax.bar(indices, expert, width=bar_width, label='Expert', color=palette[0])
ax.bar(indices + 670, single, width=bar_width, label='Single-Agent', color=palette[1])
ax.bar(indices + 1340, hi_rice, width=bar_width, label='Hi-Rice', color=palette[2])
ax.set_xlim(0, 656 + 1340)

# 添加虚线分隔各组
plt.axvline(x=663, color='black', linestyle='--')
plt.axvline(x=1333, color='black', linestyle='--')

# 自定义 y 轴
plt.ylabel('The Intervention Capacity Score \n of Each i-th Region ($RC-Q_i$)', size=22)
plt.yticks(size=20)
ax.set_yscale('log')
ax.set_yticks([1, 10, 100, 1000, 10000])
ax.get_yaxis().set_major_formatter(plt.ScalarFormatter())

# 自定义 x 轴
plt.xlabel('Region Index', size=22)
plt.xticks(
    np.concatenate([
        np.arange(0, 656, 100), 
        np.arange(670, 670 + 656, 100), 
        np.arange(1340, 1340 + 656, 100)
    ]), 
    labels=np.concatenate([
        np.arange(0, 656, 100), 
        np.arange(0, 656, 100), 
        np.arange(0, 656, 100)
    ]), 
    fontsize=20
)

# 添加分组标签
plt.text(656/2, max(np.concatenate([expert, single, hi_rice])) * 1.1, None, horizontalalignment='center', size=24, color='black')
plt.text(670 + 656/2, max(np.concatenate([expert, single, hi_rice])) * 1.1, None, horizontalalignment='center', size=24, color='black')
plt.text(1340 + 656/2, max(np.concatenate([expert, single, hi_rice])) * 1.1, None, horizontalalignment='center', size=24, color='black')

# 移除图例
plt.legend().set_visible(False)
plt.savefig("ab_region_q2.png")
plt.show()
```
